### 여러개의 이미지 자료를 한개의 배열 자료로 저장하려면
- 전제조건 : 행열의 크기가 같아야함. (resize 필수)
- 이미지를 읽음 -> resize 로 크기 통일 -> 배열로 변환 -> /255로 스케일링 -> 변환된 어레이를 저장할 리스트 변수에 추가
<br>위의 과정을 for로 작업하면 됨

- 이 과정이 다 끝나서 리스트변수에 2개 이상의 이미지 자료가 0 ~ 1의 숫자값으로 결과가 됨
- 이 결과자료를 .npy 라는 리스트 자료로 저장해서 갖고다님

- 다른 파일에서 np.load('...npy')자료를 읽어서 배열로 가져올 수 있음
- 참고 : csv 자료로 npy로 저장할 수 있음.

```python
from PIL import Image


fileList = ['./picture/pic1.jpg', './picture/pic2.jpg', './picture/pic3.jpg']
imgList = []

for fileName in fileList:
    img = Image.open(fileName)
    img = img.resize((100, 200))   # 너비(열), 높이(행)
    imgarr = np.array(img) / 255
    imgList.append(imgarr)
```

```python
# 3개의 이미지파일, 각 이미지는 200개의 행, 각 이미지는 100개의 열, 한정은 3개의 채널(rgb)
np.shape(imgList)
```

```python
## 이미지가 아닌 행열구조로의 파일로 저장하고자 하면
np.save('imgArrTransform.npy', imgList)
print('--end--')
```

```python
### 새파일이라고 가정하고
# npy로 저장된 데이터셋을 불러와서 사용가능함
# 이 자료는 이미지는 아니기 때문에 fillow 나 opencv에서 제공하는 이미지 처리 기능은 x
# 배열자료를 이미지로 저장하는 기능을 수행할 수 있음
tmp = np.load('imgArrTransform.npy')
plt.imshow(tmp[2])
```
