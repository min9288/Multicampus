### 넘파이, 판다스, 매트플랏립을 활용한 데이터 분석 실습

```python
# 그래프를 바로 보기 위한 명령어
%matplotlib inline

import numpy as np         # 넘파이 탑재
import pandas as pd        # 판다스 탑재
import matplotlib          # 매트플랏립 탑재
```

```python
# 50개의 난수로 이루어진 데이터 생성
data = np.random.rand(50)
data
```

```python
# 판다스 시리즈형 데이터로 변환
seri = pd.Series(data)
seri
```

```python
# 선형 그래프 그리기
seri.plot()
```

```python
# 그래프 스타일 변경하기
matplotlib.style.use('ggplot')
# 선형 그래프 다시 그리기
seri.plot()
```

```python
# 데이터 세트 만들기 (3개의 난수로 이루어진 10개의 배열)
data_set = np.random.rand(10, 3)
data_set
```

```python
# 데이터프레임 만들기
df = pd.DataFrame(data_set, columns=['A', 'B', 'C'])
df
```

```python
# 선형 그래프 그리기
df.plot()
```

```python
# 막대 그래프 그리기
df.plot(kind='bar')
```

```python
# 옆으로 누운 막대 그래프 그리기
df.plot(kind='barh')
```

```python
# 누적 영역 그래프 그리기
df.plot(kind='area')
```

```python
# 누적 하지 않은 영역 그래프 그리기
df.plot(kind='area', stacked=False)
```

```python
# 파이 그래프를 위한 5개의 난수 발생 및 데이터프레임 생성
seri2 = pd.Series(np.random.rand(5), index=['a', 'b', 'c', 'd', 'e'], name='series')
seri2.plot(kind='pie', autopct='%.2f', fontsize=20, figsize=(7,5))
```
